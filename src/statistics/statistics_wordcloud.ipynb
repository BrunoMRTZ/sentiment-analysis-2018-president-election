## Estatísticas perfis dos Candidatos

### Comentários no Facebook dos Candidatos

Apresentando estatísticas de todos os comentários coletados do Facebook.

#### Primeiro turno

```python
query = RawFacebookComments.select() \
        .where(
            (SQL('length(clean_comment) > 0')) & 
            (RawFacebookComments.timestamp <= datetime.date(2018, 10, 7))
        ) \
        .order_by(RawFacebookComments.data)
```

```python
words = word_frequency(query)
words_c = word_count(query)

print('Estatísticas')
print('-' * 20)
print(f'Qtde de Comentários          : {len(query)}')
print(f'Média palavras por Comentário: {round(words_c/len(query), 2)}')
print(f'Qtde de Usuários             : {len(unique_users(query))}')

create_wordcloud(query)
```
