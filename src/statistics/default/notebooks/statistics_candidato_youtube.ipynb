```python
import sys
sys.path.append('..')
sys.path.append('../..')

from stats import *

from peewee import SQL
from database.models import RawYouTubeComments
```

## Jair Bolsonaro

```python
query = RawYouTubeComments.select() \
        .where(
            (SQL('length(clean_comment) > 0')) & 
            (RawYouTubeComments.candidate == 'Jair Bolsonaro')
        ) \
        .order_by(RawYouTubeComments.hash)
```

```python
words = word_frequency(query)
words_c = word_count(query)

print('Estatísticas')
print('-' * 20)
print(f'Qtde de Comentários         : {len(query)}')
print(f'Qtde palavras por Comentário: {round(words_c/len(query), 2)}')
print(f'Qtde de Usuários            : {len(unique_users(query))}')
```

```python
create_wordcloud(query)
```
