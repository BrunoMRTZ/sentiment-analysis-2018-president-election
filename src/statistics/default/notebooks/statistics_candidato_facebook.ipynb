```python
import sys
sys.path.append('..')
sys.path.append('../..')

from stats import *

from peewee import SQL
from database.models import RawFacebookComments, RawTwitterComments, RawInstagramComments, RawYouTubeComments, RawHashtagComments
```

## Jair Bolsonaro

```python
query = RawFacebookComments.select() \
        .where(
            (SQL('length(clean_comment) > 0')) & 
            (RawFacebookComments.candidate == 'Jair Bolsonaro')
        ) \
        .order_by(RawFacebookComments.timestamp)
```

```python
dates = comments_by_date(query)
dts = list(dates.keys())
words = word_frequency(query)
words_c = word_count(query)

print('Estatísticas')
print('-' * 20)
print(f'Qtde de Comentários          : {len(query)}')
print(f'Média palavras por Comentário: {round(words_c/len(query), 2)}')
print(f'Qtde de Usuários             : {len(unique_users(query))}')
print(f'Data de coleta               : {dts[0]} <=> {dts[-1]}')
```

```python
create_wordcloud(query)
```

```python
graph_most_words_freq_by_date(RawFacebookComments, query, turno=0, candidato='Jair Bolsonaro', most_common=10)
```

```python
graph_comment_by_date(dates)
```
